```python
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import pandas as pd
import seaborn as sns
%matplotlib inline
font_path = "C:/Windows/Fonts/malgun.ttf"  
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)




# x는 구역 y는 앞바다 먼바다 중간바다
def maximum_wave(x, y):
    data = pd.read_excel('./data/combined_unique.xlsx')
    filtered_data = data[(data['region'] == x) & (data['region_sub'] == y)]
    maximum_wave_columns = '최대파고'
    maximum_avg_wave = filtered_data.groupby(['month', 'station_na'])[maximum_wave_columns].mean().unstack()
    legend_labels = maximum_avg_wave.columns[maximum_avg_wave.notna().any()].tolist()
    # 시각화
    plt.figure(figsize=(10, 6)) 
    maximum_avg_wave.plot(marker='o')
    plt.xlabel('월')
    plt.ylabel('최대파고')
    plt.legend(legend_labels, title='지점명', bbox_to_anchor=(1, 1))
    plt.grid(True)
    plt.show()
```

### 1구역 앞바다 최대파고

```python
maximum_wave(1,1)
```

### 1구역 중간바다 최대파고

```python
maximum_wave(1,2)
```

### 1구역 먼바다 최대파고

```python
maximum_wave(1,3)
```
